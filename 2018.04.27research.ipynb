```python
import pandas as pd
import numpy as np
import matplotlib as plt
from pandas import DataFrame
```

```python
data = pd.read_csv('D:/Documents/Go Abroad/实习/成交金额/Python数据/2017.csv',encoding='gbk',engine='python')
data = data.set_index('date')
data = data[['stock','close','turnover_sum']]
```

```python
L = []
for i in set(data["stock"]):
    b = list(data["stock"].values).count(i)
    L.append(b)
```

```python
a = list(set(data["stock"]))
stock_m = a[L.index(max(L))]
stock_m
```

```python
stock_m = 1
```

```python
stock_eg = data[data['stock'] == stock_m]
stock_eg
```

```python
suoyin = list(stock_eg.index)
stock_sum = data[data["stock"] == list(set(data["stock"]))[0]]
stock_sum = stock_sum.reindex(suoyin,method='ffill')
close = stock_sum['close']
change = close.pct_change(periods=1, fill_method='pad')
stock_sum["change"] = change
stock_sum
```

```python
for i in list(set(data["stock"])):
    stock_q = data[data["stock"] == i]
    stock_qb = stock_q.reindex(suoyin,method='ffill')
    close = stock_qb['close']
    change = close.pct_change(periods=1, fill_method='pad')
    stock_qb["change"] = change
    stock_sum = stock_sum.append(stock_qb)
stock_sum
```
