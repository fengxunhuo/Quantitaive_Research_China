```python
import pandas as pd
import numpy as np
import PyPDF2 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

Scrape data from PDF file

```python
## Scrape data from the PDF file into Python
filename = 'D:/Documents/University Life/UMN/Career pre/20181202Assement/Financial Regulation-2001.pdf'
pdfFileObj = open(filename,'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

```python
## Get pages of pdf
num_pages = pdfReader.numPages
count = 0
text = ""

## Read data from each page into variable "text"
while count < num_pages:
    pageObj = pdfReader.getPage(count)
    count +=1
    text += pageObj.extractText()
```

Properly transform data and save those into several tables, export in spreadsheets.

```python
## Create a string of all article number
keyword = ""
for num in range(1,25):
    keyword += "ARTICLE" + " " + str(num) + "." + ","

## Split strings according to comma
keyword = keyword.split(",")
```

```python
## Create a new dict
clause_tot = dict.fromkeys(keyword,"")
for i in range(0,23):
    clause_tot[keyword[i]] = text.split(keyword[i])[1].split(keyword[i+1])[0]

## Add the last one 
clause_tot[keyword[23]] = text.split(keyword[23])[1]
```

```python
abstract = []
clause = []
## Split ARTICLE and clause details
for article in keyword:
    abstract_i = (clause_tot[article])[0:65].split(".")[0].strip()
    abstract.append(abstract_i)
    clause_i = clause_tot[article].strip(abstract_i).lstrip(".").strip()
    clause.append(clause_i)

## Delete the last elements
abstract.pop()
clause.pop()
```

```python
## Delete redundant element
keyword.pop()
```

```python
## Create an empty dataframe
Spreadsheets = pd.DataFrame()
## Create a new list
documents = list(np.repeat("AGREEMENT No. 4-2001", 24))

## Insert columns
Spreadsheets.insert(0, "Documents", documents)
Spreadsheets.insert(1, "Agrees", keyword)
Spreadsheets.insert(2, "Abstract", abstract)
Spreadsheets.insert(3, "Clause", clause)
```

```python
## This is a function to generate a dataframe for clause details
def df_clauses(k, clause_detail_k):
    ## Create necessary lists
    documents_k = list(np.repeat("AGREEMENT No. 4-2001", len(clause_detail_k)))
    agrees_k = list(np.repeat(keyword[k-1], len(clause_detail_k)))
    abstract_k = list(np.repeat(abstract[k-1], len(clause_detail_k)))

    df_clause_k = pd.DataFrame()
    ## Insert columns
    df_clause_k.insert(0, "Documents", documents_k)
    df_clause_k.insert(1, "Agrees", agrees_k)
    df_clause_k.insert(2, "Abstract", abstract_k)
    df_clause_k.insert(3, "Clause", clause_detail_k)
    
    return df_clause_k
```
