```python
# encoding:utf-8
import talib as ta
from fxdayu_data import DataAPI
from datetime import datetime
import tushare as ts
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from itertools import chain
```

```python
start=datetime(2014,1,1)
end=datetime(2016, 12, 31)
index = ["SME", "GEM"]
codes = set(chain(*map(DataAPI.candle.find, index)))
print codes
SME = DataAPI.candle(tuple(codes), "D", start=start, end=end)
```

```python
SME
```

```python
prices = SME.minor_xs('close')
prices
```

```python
LPN = DataAPI.factor(tuple(codes), ('LCAP','PVI','NetProfitGrowRate'), start=datetime(2014,1,1), end=datetime(2016, 12, 31))
print LPN
```

```python
def winsorize(df):
    q = df.quantile([0.025, 0.975])
    if isinstance(q, pd.Series) and len(q) == 2:
        df[df < q.iloc[0]] = q.iloc[0]
        df[df > q.iloc[1]] = q.iloc[1]
    return df

def standardize(df):
    df_std = df.std()
    df_mean = df.mean()
    return (df - df_mean) / df_std

S_LCAP = standardize(winsorize(LPN.minor_xs('LCAP')))
S_PVI = standardize(winsorize(LPN.minor_xs('PVI')))
S_NPG = standardize(winsorize(LPN.minor_xs('NetProfitGrowRate')))
fig = plt.figure(figsize=(15, 7))
plt.plot(S_LCAP)
plt.show()
```

```python
S = -(3*S_LCAP+1*S_PVI+S_NPG)
factor = S.stack()

print(factor)
```

```python
import alphalens
factor_data = alphalens.utils.get_clean_factor_and_forward_returns(factor, prices, quantiles=5)
factor_data
```

```python
mean_return_by_q, std_err_by_q = alphalens.performance.mean_return_by_quantile(factor_data, by_date=True)
ic = alphalens.performance.factor_information_coefficient(factor_data)
mean_monthly_ic = alphalens.performance.mean_information_coefficient(factor_data, by_time='M')
ic
```

```python
import matplotlib.pyplot as plt
alphalens.plotting.plot_cumulative_returns_by_quantile(mean_return_by_q, 10)
alphalens.plotting.plot_cumulative_returns(mean_return_by_q.iloc[:,0], period=1)
alphalens.plotting.plot_ic_hist(ic)
alphalens.plotting.plot_monthly_ic_heatmap(mean_monthly_ic)
plt.show()
```
