```python
import numpy as np
import talib as ta
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```python
data = pd.read_csv('D:/Documents/Go Abroad/实习/成交金额/Python数据/2017_4/2017_stockid.csv',encoding='gbk',engine='python')
```

```python
stock_id = list(set(data['stock']))
stock_id.sort()
date_id = list(set(data['date']))
date_id.sort()
data_new = pd.DataFrame(index = date_id, columns = stock_id)
data_new
```

```python
for i in date_id:
    stock_a = data[data['date'] == i]['stock'].values
    for k in stock_a:
        data_new.loc[i,k] = 1  
```

```python
stock_ido = []
for n in stock_id:
    n = str(n)
    if n.startswith("60"):
        if len(n) != 6:
            n = int(n)
            a = '%06d' % n
            stock_ido.append(a+".XSHE")
        else:
            stock_ido.append(n+".XSHG")
    else:
        n = int(n)
        a = '%06d' % n
        stock_ido.append(a+".XSHE")
```

```python
data_new.columns = stock_ido
data_new = data_new.replace(np.nan, 0)
#data_new.columns.name = "asset"
data_new.index.name = "date"
data_new
```
